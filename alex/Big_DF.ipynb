```python
import os
import pandas as pd

```

```python
csv = pd.read_csv("./nonfinancial.csv") # Load our list of nonfinancial features we wanted. (I further cleaned the list up to only include a commodity. So I also removed swaps and forwards and index.)
data_dir = "C://Users/alexb/Documents/class/project3data/dump/"
 # path to where the csvs are. This could also be a quandl qu
files = os.listdir(data_dir)
lists = []
tkr_code = (csv["Quandl Code"]).values
code = []
count = 1
#makes a list of just the codes
while count < 205:
    count += 1
    code.append(tkr_code[count][6:])
```

```python
file_list = []
count = 0
valid_code= []
big_df = pd.DataFrame()
#finds everything that we had in both lists
while count < 203:
    count += 1
    if (any(code[count] in item for item in files) == True):
        valid_code.append(code[count])
    else:
        print(code[count])
```

```python
count = 0
final_df = pd.DataFrame()
while count < (len(valid_code)-1):
    count +=1
    try:
        df =pd.read_csv(data_dir+valid_code[count]+"1.csv")
        df = df[["Date", "Settle", "Volume"]]
        df = df.rename(columns={"Settle": valid_code[count]+"S", "Volume": valid_code[count]+"V", "Date": valid_code[count]+"D"})
        final_df = pd.concat([final_df, df], axis=1)
    except:
        print(f"{valid_code[count]} not added")
```

```python
final_df.to_csv("features.csv")
```
